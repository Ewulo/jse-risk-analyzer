```python
# import libraries
import yfinance as yf
import pandas as pd
import os
```

```python
tickers = {
    "Naspers":        "NPN.JO",
    "MTN Group":      "MTN.JO",
    "Standard Bank":  "SBK.JO",
    "Anglo American": "AGL.JO",
    "Shoprite":       "SHP.JO"
}
START = "2019-01-01"
END   = "2024-12-31"
```

```python
raw_data = yf.download(list(tickers.values()), start=START, end=END, auto_adjust=True)
```

```python
prices = raw_data["Close"]

# Rename columns from tickers to company names for readability
prices.columns = list(tickers.keys())

# Drop rows where ALL prices are missing
prices.dropna(how="all", inplace=True)
```

```python
# shape and preview
print("Shape:", prices.shape)
print("\nFirst 5 rows:")
print(prices.head())
```

```python
print("Missing values per stock:\n")
print(prices.isnull().sum())
```

```python
# Save to CSV
os.makedirs("../data", exist_ok=True)
prices.to_csv("../data/jse_prices.csv")
print("\n✅ Prices saved to data/jse_prices.csv")
```
